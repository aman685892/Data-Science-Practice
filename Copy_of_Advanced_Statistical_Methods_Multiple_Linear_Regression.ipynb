```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data=pd.read_csv('/content/1.02.+Multiple+linear+regression (1).csv')
data
```

```python
data.describe()
```

**FIRST MULTIPLE REGRESSION**

```python
y=data['GPA']
x2=data[['SAT','Rand 1,2,3']]
```

```python
x=sm.add_constant(x2)
results=sm.OLS(y,x).fit()
print(results.summary())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data=pd.read_csv('/content/1.03.+Dummies.csv')
raw_data
```

```python
data=raw_data.copy()
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
```

```python
##REGRESSION
```

```python
y=data['GPA']
x3=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x3)
results=sm.OLS(y,x).fit()
print(results.summary())
```

```python
plt.scatter(data['SAT'],y)
yhat_No=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
fig=plt.plot(data['SAT'],yhat_No, lw=2,c='red')
fig=plt.plot(data['SAT',yhat_yes, lw=2,c='green'])
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
