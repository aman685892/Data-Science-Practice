```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#Load the data
data = pd.read_csv('/1.01.+Simple+linear+regression.csv')

data
```

```python
data.describe()
```

```python
# creating GPA based on SAT exam
```

CREATING FIRST REGRESSION


```python
y=data['GPA']
x1=data['SAT']
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize=20)  # EXPLORE THE DATA
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
# REGRESSION ITSELF
x= sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1+0.2750
fig=plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# load the data
data=pd.read_csv('/real_estate_price_size.csv')
data.describe()
```

```python
y=data['price']
x2=data['size']
plt.scatter(y,x2)
plt.xlabel('size',fontsize=30)
plt.ylabel('price',fontsize=30)
plt.show()
```

```python
# REGRESSION ITSELF
x=sm.add_constant(x2)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x2,y)
yhat=1.019e+05+223.1787*x2
fig=plt.plot(x2,yhat,lw=5,c='red',label='regression line')
plt.xlabel('size',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.show()
```

BEAUTY OF SEABORN FOR GRAPHS

```python
plt.scatter(x2,y)
yhat=0+223.1787*x2
fig=plt.plot(x2,yhat,lw=5,c='green',label='regression line')
plt.xlabel('size',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```
